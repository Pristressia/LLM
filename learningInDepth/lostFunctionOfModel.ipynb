### Loss function ทำงานตรงกับ Logits มาก 
Cross-entropy + softmax มี gradient : 
$$\frac {\partial \mathcal {L}} {\partial logits} = p - y$$

่Where  
**logits** คือ คะแนนดิบที่ได้จากโมเดล (หลังผ่านบล็อค Linear ก่อนเข้าสมการ Softmax)
 * ขนาด = vocab size
 * ยังไม่ใช่ probability


ยกตัวอย่าง

กำหนดให้ vocab มีทั้งหมด 3 คำ
โมเดลคำนวนได้ $$logits = [2.0, 4.0, 1.0]$$

จะได้ว่ามีโอกาสที่จะออก token ที่ index = 1 มากที่สุดแต่นี่ไม่ใช่ค่าความน่าจะเป็น

**$p$ คือค่า probability** ที่จะได้มาจากการนำ logits เข้าไปใส่ในฟังก์ชัน Softmax

$$p = softmax(logits)$$

เช่น

$$p = [0.11, 0.84, 0.05]$$

**$y$ คือคำตอบจริง (ground truth)** ในรูปแบบ one-hot vector 

* ถ้าค่าที่ถูกต้องคือ "dog" (index = 1):

$$y = [0, 1, 0]$$


 ### การควบคุมการ generate ได้ (temperature, top-k)

temperature increase => high confidential
temperature decrease => more randomly

$$logits^\prime = \frac {logits} {temperature}$$

all happen at logits level

### อธิบายเพิ่มในส่วนของ Temperature 

from real LLM pipeline :
$$logits \rightarrow ( / temperature ) \rightarrow softmax \rightarrow loss$$
can write it to function 

$$\mathcal {L} = \mathcal {L} \left( softmax \left( \frac {logits}{T} \right)\right)$$

ดังนั้นในกระบวนการ back propagation:
$$\frac {\partial \mathcal {L}} {\partial\ logits} = \frac {\partial \mathcal {L}} {\partial logits^\prime} \cdot \frac {\partial logits^\prime} {\partial logits}$$

สรุป Temperature คือค่าที่ถูกนำมาใช้เพื่อ scale ค่า logits ก่อนโยนเข้า softmax
